```python
def da_RFE_find_features(preprocessor,classifier,X_train,y_train,smote_type='nc',scoring='recall'):
    '''Finds best number of features to keep in the dataset. 
    
    Parameters
    ---------------
    preprocessor (ColumnTransformer): Transformer to apply to data
    classifier (object): Type of estimator to use, e.g. LogisticRegression()
    X_train, y_train (DataFrames): XY data to use in calculations.
    smote type, default nc
    scoring type for the cross validation process to return. Default recall

    Returns
    ---------------
    Visualization of best number of features to keep, dataframe of results in rank order
    '''
    
    if smote_type.lower()=='nc':
        smote=SMOTENC(categorical_features=X_train.dtypes==object,random_state=1234,n_jobs=-1)
        X_res,y_res=smote.fit_resample(X_train,y_train)
    else:
        X_res=X_train
        y_res=y_train
        
    X_train_scaled=preprocessor.fit_transform(X_res,y_res)

    feature_names=preprocessor.get_feature_names_out()
    X_train_scaled=pd.DataFrame(data=X_train_scaled,columns=feature_names)

    cv = StratifiedKFold(5)
    visualizer = rfecv(classifier, X=X_train_scaled, y=y_res, cv=cv, scoring=scoring)
        
    col_ranks=pd.DataFrame(columns=["col",'rank'])
    col_ranks['col']=X_train_scaled.columns
    col_ranks['rank']=visualizer.ranking_
    
    
    return visualizer,col_ranks.sort_values(by='rank')
```
